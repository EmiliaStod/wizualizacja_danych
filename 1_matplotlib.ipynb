### Histogram

Histogram przedstawia rozłożenie różnych wartości w danej kategorii. Jego wygląd jest podobny do wykresu kolumnowego, jednak jego zastosowanie jest zupełnie inne i wykresy te nie powinny być ze sobą mylone.

Przygotuj histogram (`hist`) opierając się na danych, które sam wybierzesz w sensowny sposób. Zapoznaj się z parametrami `bins` i `range` oraz dobierz dla nich takie wartości, które wg. Ciebie najlepiej wizualizują dane. Dodatkowo, zapoznaj się z parametrem `density`. Zadbaj o czytelność wykresu (tytuł wykresu oraz podpisy osi). 

Histogram porównujący wzrost kobiet i mężczyzn z użyciem density aby zobaczyć rozkład danych

```python
import pandas as pd
import matplotlib.pyplot as plt

# Wczytanie danych
df = pd.read_csv('athlete_events_pd.csv')

# Podział na płeć
height_men = df[df['Sex'] == 'M']['Height']#Wybieramy tylko wiersze, które spełniają warunek (True) i wybieramy kolumnę 'Height'
height_women = df[df['Sex'] == 'F']['Height']

# Parametry histogramu
bins = 20  # Liczba przedziałów
range_min, range_max = 140, 220  # Ustalony zakres wzrostu od 140 cm do 220 cm

# Tworzenie histogramu wzrostu kobiet i mężczyzn
plt.figure(figsize=(12, 6))
plt.hist(height_men, bins=bins, range=(range_min, range_max), alpha=0.5, color='blue', label='Mężczyźni', density=True)
plt.hist(height_women, bins=bins, range=(range_min, range_max), alpha=0.5, color='red', label='Kobiety', density=True)

# Opisy osi i tytuł
plt.xlabel('Wzrost (cm)', fontsize=12)
plt.ylabel('Gęstość prawdopodobieństwa', fontsize=12)
plt.title('Porównanie wzrostu sportowców (kobiet i mężczyzn), którzy brali udział w wydarzeniach sportowych', fontsize=14)
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Wyświetlenie wykresu
plt.show()
```

##### ⭐ Zadanie 3:

Przygotuj dwuwymiarowy histogram z heksagonami (`hexbin`) opierając się na danych, które sam wybierzesz w sensowny sposób. Zapoznaj się z parametrem `gridsize` oraz dobierz dla niego taką wartość, która wg. Ciebie najlepiej wizualizuje dane. Zadbaj o czytelność wykresu (tytuł wykresu, podpisy osi oraz legenda). 

```python
df = df.dropna(subset=['Weight', 'Height'])#usunięcie wierszy z brakującymi danymi

height_men = df[df['Sex'] == 'M']['Height']
weight_men = df[df['Sex'] == 'M']['Weight']

# Tworzenie wykresu hexbin dla Waga vs. Wiek
plt.figure(figsize=(12, 6))
plt.hexbin(height_men, weight_men, gridsize=50, cmap='inferno', mincnt=1)

# Dostosowanie osi X
plt.xlim(120, 240) # Zakres od 120 do 240 cm
plt.xticks(range(120, 241, 10))  # Co 10 cm

# Opisy osi i tytuł
plt.xlabel('Wzrost (cm)', fontsize=12)
plt.ylabel('Waga (kg)', fontsize=12)
plt.title('Histogram ukazujący zależność wzrostu sportowców płci męskiej do ich wagi', fontsize=14)
cbar = plt.colorbar()
cbar.set_label('Liczba sportowców')

# Wyświetlenie wykresu
plt.show()
```
